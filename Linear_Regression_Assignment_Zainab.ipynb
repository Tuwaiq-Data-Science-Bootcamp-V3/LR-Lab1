## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.model_selection import train_test_split


X = bostonData.loc[:, ~bostonData.columns.isin(['Price'])] #X is the dataframe columns except the price column
y = bostonData.Price #the target y is the price column

#split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=23)
```

I use this code from [statology](https://www.statology.org/pandas-exclude-column/#:~:text=You%20can%20use%20the%20following,columns!%3D') to Exclude Price column.

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

## Q9: Train the Model.

```python
model.fit(X_train, y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(f"intercept: \n\n{model.intercept_}")
```

```python
print(f"slope: \n\n{model.coef_}")
```

```python
y_pred = model.predict(X_test)
```

```python
print("y_test: \n\n",np.array(y_test))
```

```python
print("y_pred: \n\n",np.array(y_pred))
```

```python
plt.scatter(x=y_test, y=y_pred)
plt.title('Relation between y_test and y_pred')
plt.show()
```

## Residual Histogram

```python
# y here is an arbitrary target variable for explaining this example    
residuals = y_test - y_pred 

import seaborn as sns
sns.distplot(residuals) # you may select the no. of bins
plt.title('Error Terms', fontsize=20)           
plt.xlabel('Residuals', fontsize = 15)     
plt.show()
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
import sklearn.metrics as metrics
```

```python
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(y_test,y_pred)
```

```python
print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
```

```python
print("R^2: ", r2)
```
