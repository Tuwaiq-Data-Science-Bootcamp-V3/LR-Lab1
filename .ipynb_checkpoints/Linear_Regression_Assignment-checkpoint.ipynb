## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.model_selection import train_test_split
```

```python
x= df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y= df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y , test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

## Q9: Train the Model.

```python
lr.fit(X_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(lr.intercept_)
```

```python
coffe_df= pd.DataFrame(lr.coef_, x.columns, columns=['Coefficient'])
coffe_df
```

```python
predictions = lr.predict(X_test)
predictions
```

```python
Real_Values = np.array(y_test)
Real_Values
```

```python
# Ploting them to see the relationship between the real Values & predictions
plt.scatter(Real_Values,predictions);
```

## Residual Histogram

```python
sns.distplot((y_test-predictions), bins=50);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics
from sklearn.metrics import r2_score
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
print('R**2:', r2_score(y_test, predictions))
```
