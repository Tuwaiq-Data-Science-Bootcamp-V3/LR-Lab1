## Q4: Use a heatmap to show the correlation between features and the target labels.

```python
# write your code here ^_^

# Create a correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix as a heatmap with the specified color palette
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.model_selection import train_test_split
X = boston.data
```

```python
y = boston.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression
# Create an instance of Linear Regression model
model = LinearRegression()
```

## Q9: Train the Model.

```python
# Fit the model to the training data
model.fit(X_train, y_train)
```

```python

# Make predictions on the test data
y_pred = model.predict(X_test)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
# Print the intercept and coefficients
print("Intercept:", model.intercept_)
```

```python
print("Coefficients:")
for feature_name, coef in zip(boston.feature_names, model.coef_):
    print(feature_name, ":", coef)
    
```

```python
# Make predictions on the test data
y_pred = model.predict(X_test)
y_pred
```

```python
a_p=np.array(df['Price'])
a_p
```

```python
# Scatter plot of predictions vs. real values
plt.scatter(y_test, y_pred)
plt.xlabel('Real Values')
plt.ylabel('Predictions')
plt.title('Predictions vs. Real Values')
plt.show()
```

## Residual Histogram

```python
residuals = y_test - y_pred

# Plot histogram of residuals
sns.distplot(residuals, bins=20)
plt.xlabel('Price')
plt.ylabel('Density')
plt.title('Histogram of Residuals')
plt.show()
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Print the evaluation metrics
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)
```
