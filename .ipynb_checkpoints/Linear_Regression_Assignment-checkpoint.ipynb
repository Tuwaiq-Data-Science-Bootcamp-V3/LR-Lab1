## Import Libraries

```python
# To ignore warnings
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Our Goal: to predict the price given a set of attributes.

## Load the Data

```python
boston = pd.read_csv("df_boston.csv")
boston
```

```python
boston.columns
```

```python
boston = boston.drop('Unnamed: 0', axis=1)
```

```python
X = boston[['CRIM', 'ZN', 'INDUS','CHAS', 'NOX','RM','AGE','DIS', 'RAD' ,'TAX' ,'PTRATIO', 'B', 'LSTAT']]
y = boston['Price']

X_boston,y_boston = X, y
print('Shape of data:', X_boston.shape, y_boston.shape)
```

## EDA (Exploratory Data Analysis)

## Q7: Split the dataset into Train and Test sets with test_size=30% and random_state=23.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_boston,y_boston, test_size=0.3, random_state=23)
```

## Q8: Build a Linear Regression Model.

```python
from sklearn.linear_model import LinearRegression

LG = LinearRegression()
```

## Q9: Train the Model.

```python
LG.fit(X_train,y_train)
```

## Q10: Evaluate the model. 
- print intercept and coefficients.
- compare between predictions and real values, then visualize them.
- Draw Residual Histogram.

```python
print(LG.intercept_) 
```

```python
coeff_df = pd.DataFrame(LG.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

## Predictions vs. Real Values

```python
predictions = LG.predict(X_test)
predictions
```

```python
Real_Values = np.array(y_test)
Real_Values
```

```python
plt.scatter(Real_Values,predictions);
```

## Residual Histogram

```python
sns.distplot((y_test-predictions),bins=20);
```

## Q11: Use evaluation metrics MAE, MSE, RMSE and R^2.

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
print('R^2:', r2_score(y_test,predictions))
```
